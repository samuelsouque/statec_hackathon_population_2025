```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('../../data/life_expectancy.csv')
df = df[['Sex', 'TIME_PERIOD', 'OBS_VALUE']]
df_retirement = pd.read_csv('../../data/retairement_age.csv')
```

```python
olivia_df = pd.read_excel('../../data/hakathon 3.xlsx', sheet_name = 1)
```

```python
olivia_df.columns
```

```python
result = {}
years = np.union1d(df['TIME_PERIOD'].unique(), olivia_df['Year'].unique())

for year in years:
    olivia_df_year = olivia_df[olivia_df['Year'] == year]
    df_year = df[df['TIME_PERIOD'] == year]
    df_retirement_year = df_retirement[df_retirement['year'] == year]

    # Retirement age
    if not olivia_df_year.empty and 'Pension Age' in olivia_df_year.columns:
        retirement_age = olivia_df_year['Pension Age'].iloc[0]
    elif not df_retirement_year.empty and 'retirement_age' in df_retirement_year.columns:
        retirement_age = df_retirement_year['retirement_age'].iloc[0]
    else:
        retirement_age = None

    # Male life expectancy
    if not df_year[df_year['Sex'] == 'Men'].empty:
        life_m = df_year[df_year['Sex'] == 'Men']['OBS_VALUE'].iloc[0]
    elif not olivia_df_year.empty and 'M Life expectancy' in olivia_df_year.columns:
        life_m = olivia_df_year['M Life expectancy'].iloc[0]
    else:
        life_m = None

    # Female life expectancy
    if not df_year[df_year['Sex'] == 'Women'].empty:
        life_f = df_year[df_year['Sex'] == 'Women']['OBS_VALUE'].iloc[0]
    elif not olivia_df_year.empty and 'F Life expectancy' in olivia_df_year.columns:
        life_f = olivia_df_year['F Life expectancy'].iloc[0]
    else:
        life_f = None

    # Build result
    result[str(year)] = {
        'population': olivia_df_year['Total Population'].iloc[0] if not olivia_df_year.empty else None,
        'beneficiary_avg': olivia_df_year['N of pensions'].iloc[0] if not olivia_df_year.empty else None,
        'life_m': life_m,
        'life_f': life_f,
        'retirement_age': retirement_age
    }
```

```python
with open('../../application/data.json', 'w') as f:
    json.dump(result, f, indent=4)
```
